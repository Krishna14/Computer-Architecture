```python
import keras
def ConvNetwork(x_train,x_test):
    input_img = Input(shape=(256,256,3))
    """This function generates the convolutional neural network"""
    #Initial size is 1 x 28 x 28
    x = Conv2D(16, (5, 5), activation='relu', padding='same')(input_img)
    #New size is 16 x 28 x 28
    x = MaxPooling2D((2, 2), padding='same')(x)
    #New size is 16 x 14 x 14
    x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
    #New size is 8 x 14 x 14
    x = MaxPooling2D((2, 2), padding='same')(x)
    #New size is 8 x 7 x 7
    x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
    #New size is 8 x 7 x 7
    encoded = MaxPooling2D((2, 2), padding='same')(x)
    #New size is 8 x 4 x 4
    
    # at this point the representation is (4, 4, 8) i.e. 128-dimensional
    
    x = Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
    ##New size is 8 x 4 x 4 
    x = UpSampling2D((2, 2))(x)
    ##New size is 8 x 8 x 8
    x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
    ##New size is 8 x 8 x 8
    x = UpSampling2D((2, 2))(x)
    ##New size is 8 x 16 x 16
    x = Conv2D(16, (3, 3), activation='relu', padding='same')(x)
    ##New size is 16 x 14 x 14
    x = UpSampling2D((2, 2))(x)
    ##New size is 16 x 28 x 28
    decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)
    ##New size is 1 x 28 x 28
    
    autoencoder = Model(input_img, decoded)
    Adam = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
    autoencoder.compile(optimizer=Adam, loss='mean_squared_error')
    autoencoder.fit(x_train, x_train, epochs=3, batch_size=128, shuffle=True, validation_data=(x_test, x_test))
    keras.backend.clear_session()
    del autoencoder
```
